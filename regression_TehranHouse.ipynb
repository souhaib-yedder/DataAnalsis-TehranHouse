```python
#تصحيج للكود السابق 
import pandas as pd

# قراءة البيانات من الملف
df = pd.read_csv('TehranHouse.csv')

# الحصول على القيم الفريدة من عمود Address
unique_addresses = df['Address'].unique()

# إنشاء DataFrame جديد للقيم الفريدة
unique_df = pd.DataFrame(unique_addresses, columns=['Unique_Address'])

# إضافة عمود الأرقام من 1 إلى عدد العناصر الفريدة
unique_df['Address_Number'] = range(1, len(unique_df) + 1)

# إنشاء عمود لتمثيل كل عنصر في العمود الثالث
df['Address_Representation'] = df['Address'].map(dict(zip(unique_df['Unique_Address'], unique_df['Address_Number'])))

# تحويل عمود Area إلى نوع كسري (float)
df['Area'] = df['Area'].str.replace(' ', '')  # إزالة الفراغات
df['Area'] = df['Area'].str.replace(',', '')  # إزالة الفواصل
df['Area'] = pd.to_numeric(df['Area'], errors='coerce')  # تحويل إلى نوع رقمي

# عرض DataFrame الجديد مع الأعمدة الجديدة
print(unique_df)

# عرض بعض الصفوف فقط من DataFrame الأصلي مع العمود Area بعد التحويل
print(df[['Address', 'Address_Representation', 'Area']].head())  # عرض أول 5 صفوف
```

```python


# افترض أن لديك DataFrame df بالفعل

# تحويل الأعمدة من نوع bool إلى نوع int
bool_columns = ['Parking', 'Warehouse', 'Elevator']  # استبدل هذه الأسماء بأسماء الأعمدة الخاصة بك
df[bool_columns] = df[bool_columns].astype(int)

# تحديد الأعمدة العددية (بما في ذلك الأعمدة المحولة)
df_numeric = df.select_dtypes(include=['number'])

# حساب مصفوفة الارتباط
df_corr = df_numeric.corr()

# رسم خريطة الحرارة
plt.figure(figsize=(12, 8))
sns.heatmap(df_corr, cmap='Reds', annot=True, fmt=".2f", square=True, cbar_kws={"shrink": .8})

# إضافة عنوان
plt.title('Feature Correlation Matrix')

# عرض الخريطة
plt.show()
```

```python


# Assume you have a DataFrame df already loaded with data

# Plotting the relationship between Price and other features
plt.figure(figsize=(12, 8))

# First plot: Price vs Room
plt.subplot(2, 3, 1)
plt.scatter(df['Room'], df['Price'], c='blue', alpha=0.5)
plt.xlabel('Number of Rooms (Room)')
plt.ylabel('Price')
plt.title('Relationship between Number of Rooms and Price')

# Second plot: Price vs Parking
plt.subplot(2, 3, 2)
plt.scatter(df['Parking'], df['Price'], c='green', alpha=0.5)
plt.xlabel('Parking Availability')
plt.ylabel('Price')
plt.title('Relationship between Parking and Price')

# Third plot: Price vs Warehouse
plt.subplot(2, 3, 3)
plt.scatter(df['Warehouse'], df['Price'], c='orange', alpha=0.5)
plt.xlabel('Warehouse Availability')
plt.ylabel('Price')
plt.title('Relationship between Warehouse and Price')

# Fourth plot: Price vs Elevator
plt.subplot(2, 3, 4)
plt.scatter(df['Elevator'], df['Price'], c='red', alpha=0.5)
plt.xlabel('Elevator Availability')
plt.ylabel('Price')
plt.title('Relationship between Elevator and Price')

# Fifth plot: Price vs Area
plt.subplot(2, 3, 5)
plt.scatter(df['Area'], df['Price'], c='purple', alpha=0.5)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Relationship between Area and Price')

# Sixth plot: Price vs Address_Representation
plt.subplot(2, 3, 6)
plt.scatter(df['Address_Representation'], df['Price'], c='brown', alpha=0.5)
plt.xlabel('Address Representation')
plt.ylabel('Price')
plt.title('Relationship between Address Representation and Price')

# Improve layout
plt.tight_layout()
plt.show()

# Print a message indicating that the plots have been displayed
print("All relationship plots between features and Price have been displayed.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume you have a DataFrame df already loaded with data

# Setting up the figure for the histograms
plt.figure(figsize=(12, 8))

# Histogram for Price vs Room
plt.subplot(2, 3, 1)
plt.hist(df[df['Room'] == 1]['Price'], bins=20, alpha=0.5, label='1 Room', color='blue')
plt.hist(df[df['Room'] == 2]['Price'], bins=20, alpha=0.5, label='2 Rooms', color='green')
plt.hist(df[df['Room'] == 3]['Price'], bins=20, alpha=0.5, label='3 Rooms', color='orange')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution by Number of Rooms')
plt.legend()

# Histogram for Price vs Parking
plt.subplot(2, 3, 2)
plt.hist(df[df['Parking'] == 1]['Price'], bins=20, alpha=0.5, label='With Parking', color='green')
plt.hist(df[df['Parking'] == 0]['Price'], bins=20, alpha=0.5, label='Without Parking', color='red')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution by Parking Availability')
plt.legend()

# Histogram for Price vs Warehouse
plt.subplot(2, 3, 3)
plt.hist(df[df['Warehouse'] == 1]['Price'], bins=20, alpha=0.5, label='With Warehouse', color='orange')
plt.hist(df[df['Warehouse'] == 0]['Price'], bins=20, alpha=0.5, label='Without Warehouse', color='purple')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution by Warehouse Availability')
plt.legend()

# Histogram for Price vs Elevator
plt.subplot(2, 3, 4)
plt.hist(df[df['Elevator'] == 1]['Price'], bins=20, alpha=0.5, label='With Elevator', color='red')
plt.hist(df[df['Elevator'] == 0]['Price'], bins=20, alpha=0.5, label='Without Elevator', color='brown')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution by Elevator Availability')
plt.legend()

# Histogram for Price vs Area
plt.subplot(2, 3, 5)
plt.hist(df['Area'], bins=20, alpha=0.5, color='purple')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.title('Area Distribution')

# Histogram for Price vs Address_Representation
plt.subplot(2, 3, 6)
plt.hist(df['Address_Representation'], bins=20, alpha=0.5, color='gray')
plt.xlabel('Address Representation')
plt.ylabel('Frequency')
plt.title('Address Representation Distribution')

# Improve layout
plt.tight_layout()
plt.show()

# Print a message indicating that the plots have been displayed
print("All histogram plots have been displayed.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))  # تحديد حجم الرسم

# رسم النقاط للعقارات التي لديها مواقف سيارات
plt.scatter(
    'Area', 
    'Price(USD)', 
    data=df[df['Parking'] == 1],  # افتراض أن 1 تعني وجود مواقف سيارات
    marker='x', 
    c='magenta', 
    label='With Parking'
)

# رسم النقاط للعقارات التي ليس لديها مواقف سيارات
plt.scatter(
    'Area', 
    'Price(USD)', 
    data=df[df['Parking'] == 0],  # افتراض أن 0 تعني عدم وجود مواقف سيارات
    marker='o', 
    c='blue', 
    label='Without Parking'
)

# إضافة وسيلة إيضاح
plt.legend(title='Parking Availability')

# إضافة عناوين المحاور
plt.xlabel('Area')
plt.ylabel('Price (USD)')

# عرض الرسم
plt.title('Price by Area based on Parking Availability')
plt.show()
```
